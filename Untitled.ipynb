```python
import requests
import json
import boto3
import time
from IPython.display import JSON
import pandas as pd 
import os
from io import StringIO
from dateutil.relativedelta import relativedelta


from sqlalchemy import create_engine
engine = create_engine('mysql+pymysql://root:root@localhost/mf_reports', echo=False)

def month_range():
    result = []
    today = datetime.date.today()
    today = datetime.date(2019, 12, 1)
    current = datetime.date(2019, 1, 1)    

    while current <= today:
        result.append([current, ((current + relativedelta(months=1))-relativedelta(days=1))])
        current += relativedelta(months=1)
    return result

    
def download_history():
    months = month_range()
    for m in months:
        url = "http://portal.amfiindia.com/DownloadNAVHistoryReport_Po.aspx?frmdt={}&todt={}".format(m[0].strftime('%d-%b-%Y'),m[1].strftime('%d-%b-%Y'))
        path = "{}.txt".format(m[0].strftime('%Y%m'))
        if os.path.isfile(path):
            continue 
        r = requests.get(url)  
        with open(path, 'wb') as f:
            f.write(r.content)
            
def download_history():
    months = month_range()
    for m in months:
        url = "http://portal.amfiindia.com/DownloadNAVHistoryReport_Po.aspx?frmdt={}&todt={}".format(m[0].strftime('%d-%b-%Y'),m[1].strftime('%d-%b-%Y'))
        data = Amfi.get_nav_history(url)
        
        
def convert_to_csv():
    files = []
    with os.scandir('amfitxt') as i:
        for entry in i:
            if entry.is_file() and ".txt" in entry.name:
                files.append(entry.name)

    for file in files:
        with open('amfitxt/'+file) as f:
            amfi = Amfi()
            amfi.txt = f.read()
            schemes = amfi.get_nav_history()
            df = amfi.get_df(schemes)
            
        if not df.empty:
            name = file.replace('.txt', '.csv')
            df.to_csv("amficsv/"+name)

class Amfi():
    
    # Return txt
    def get_txt_amfi(self, url):
        resp = requests.get(url)
        if resp.status_code == 200:
            return resp.text
        else:
            print("Error in request", resp.status_code, resp.reason)
    
    # Return Array of Dict
    def get_nav_daily(self):
        url = "https://www.amfiindia.com/spages/NAVAll.txt"
        txt = self.get_txt_amfi(url)
        self.schemes = []
        lines = txt.split("\r\n")
        for line in lines:
            if len(line) > 0 and line[0] >= '0' and line[0] <= '9':
                items = line.split(";")
                if items[1].strip('-'):
                    self.schemes.append({
                        'code': items[0],'isin':items[1], 'name': items[3], 'nav': items[4], 'date': items[5]
                    })
                if items[2].strip('-'):
                    self.schemes.append({
                        'code': items[0],'isin':items[2],'name': items[3], 'nav': items[4], 'date': items[5]
                    })
        return self
    
    def get_nav_history(self, url):
        self.schemes = []
        lines = txt.split("\n")
        for line in lines:
            if len(line) > 0 and line[0] >= '0' and line[0] <= '9':
                items = line.split(";")
                
                if items[2].strip():
                    self.schemes.append({
                            'code': items[0],'isin':items[2], 'name': items[1], 'nav': items[4], 'date': items[7]
                    })
                if items[3].strip():
                    self.schemes.append({
                        'code': items[0],'isin':items[3], 'name': items[1], 'nav': items[4], 'date': items[7]
                    })
        return self.schemes
    
    def to_df(self):
        if not self.schemes:
            raise Exception("self.scheme is empty")
        
        return pd.DataFrame(self.schemes)
    
    def get_schemes(self):
        url = "http://portal.amfiindia.com/DownloadSchemeData_Po.aspx?mf=0"
        txt = self.get_txt_amfi(url)
        df = pd.read_csv(StringIO(txt))
        self.schemes = df.to_dict('records')
        
        return self.schemes
    
    def save_schemes(self)
        if not self.schemes:
            raise Exception("self.scheme is empty")
        
        df = pd.DataFrame(self.schemes, columns=["code","date","nav"])
        df['date'] =  pd.to_datetime(df['date']).dt.strftime('%Y-%m-%d') #pd.to_datetime(df['date'], format='%M-%b-%Y')
        for index, row in df.iterrows():
            query = """INSERT INTO amfi_nav (code, nav, date) VALUES(%s, %s, %s)
                ON DUPLICATE KEY UPDATE code=%s, nav=%s, date=%s"""
            engine.execute(query, (df.code[index], df.nav[index], df.date[index], df.code[index], df.nav[index], df.date[index]))
        return self.schemes
    
    
    def save(self):
        if not self.schemes:
            raise Exception("self.scheme is empty")
        
        df = pd.DataFrame(self.schemes, columns=["code","date","nav"])
        df['date'] =  pd.to_datetime(df['date']).dt.strftime('%Y-%m-%d') #pd.to_datetime(df['date'], format='%M-%b-%Y')
        for index, row in df.iterrows():
            query = """INSERT INTO amfi_nav (code, nav, date) VALUES(%s, %s, %s)
                ON DUPLICATE KEY UPDATE code=%s, nav=%s, date=%s"""
            engine.execute(query, (df.code[index], df.nav[index], df.date[index], df.code[index], df.nav[index], df.date[index]))
        return self.schemes
    
amfi = Amfi().get_nav_daily().save()
amfi
```
