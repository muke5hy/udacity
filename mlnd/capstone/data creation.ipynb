```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn  as sns
from IPython.display import display
import gc

from alpha_vantage.timeseries import TimeSeries
from alpha_vantage.techindicators import TechIndicators
from alpha_vantage.sectorperformance import SectorPerformances
from alpha_vantage.cryptocurrencies import CryptoCurrencies
import matplotlib
import matplotlib.pyplot as plt
import os
from datetime import datetime
from pytz import timezone
from time import sleep
import pickle
from datetime import date
import random
import urllib
import os

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (20.0, 10.0)

def change_timezone(date):
    newyork = timezone('America/New_York').localize(datetime.strptime(date, "%Y-%m-%d %H:%M:%S"))
    return newyork.astimezone(timezone('Asia/Kolkata')).replace(tzinfo=None)

class Data:
    def __init__(self):
        self.datadir = "/data/stocks/"
        
    def save(self, df, symbol, frequency ):
        file = f"{self.datadir}{symbol}_{frequency}.parquet.gzip"
        old = pd.DataFrame()
        if(os.path.exists(file)):
            old = pd.read_parquet(file)
            df = pd.concat([df, old])
            
        df.drop_duplicates(keep='first',inplace=True)
        df.reset_index(inplace=True, drop=True)
        df.to_parquet(file, compression='gzip')
        
    def get(self, symbol, frequency):
        file = f"{self.datadir}{symbol}_{frequency}.parquet.gzip"
        return pd.read_parquet(file)
    
    def save_pickle(self, filename, data):
        with open(f'{self.datadir}{filename}.pickle', 'wb') as handle:
            pickle.dump(data, handle, protocol=pickle.HIGHEST_PROTOCOL)
            
    def get_pickle(self, filename):
        if not os.path.exists(f'{self.datadir}{filename}.pickle'):
            return
        
        with open(f'{self.datadir}{filename}.pickle', 'rb') as handle:
            b = pickle.load(handle)
        return b
    
def read_in_chunks():
    allSymbol = []
    for chunk in pd.read_csv(datadir+'historical_minute.csv', chunksize=100000):
        allSymbol.append(chunk.tradingsymbol.unique())
def get_data(symbol):
    symbol = urllib.parse.quote(symbol)
    ts = TimeSeries(key=os.environ['ALPHA_VANTAGE_API_KEY'], output_format='pandas',indexing_type='integer')
    return ts.get_intraday(symbol=f'NSE:{symbol}',interval='1min', outputsize='full')

def can_download(symbol: str) -> bool:
    today = date.today()
    d1 = today.strftime("%Y%m%d")

    data = Data().get_pickle("download_cache")
    if not data: return True
    if not data.get(f'{symbol}{d1}'): return True
    return False


def download_nifty200():
    nifty200 = pd.read_csv('/data/stocks/master/NSE_nifty200list.csv')
    
    for symbol in nifty200['Symbol'].values:
        if not can_download(symbol):
            continue
        try:
            data, meta = get_data(symbol)
            data.rename(columns={"1. open": "open", "2. high": "high","3. low":"low","4. close":"close","5. volume":"volume"}, inplace=True)
            data.date = data.date.apply(change_timezone)
            Data().save(data,symbol,'1min')

            cache = {} if not Data().get_pickle('download_cache') else Data().get_pickle('download_cache')
            today = date.today()
            d1 = today.strftime("%Y%m%d")
            cache[f'{symbol}{d1}'] = True
            Data().save_pickle("download_cache", cache)
            sleep(random.uniform(2.0,6.))
            
        except Exception as error:
            print(symbol, error)
            sleep(random.uniform(60.0,61.))

```
